````python
#1.4

import numpy as np
import matplotlib.pyplot as plt

#création des vecteurs X et Y pour le tracé
x = []
y = []
for dico in l_points :
    x.append(dico['x'])
    y.append(dico['y'])
X = np.array(x)
Y = np.array(y)

#tracé des courbes

plt.plot(X,-Y) #plt.plot(X,Y) renvoie la courbe à l'envers
plt.axis('equal')
plt.figure()
plt.show()
````
